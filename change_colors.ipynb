# IMAGE PROCESSING

## This notebook is ti try different ways to change the color of the image and improve yolo performance

### **Initial parameters**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
TESTING = False # For testing downsample the image for faster execution
```

```python
if TESTING:

    img = cv2.imread('/home/noumena/Documents/VideoProcessing_improvingYOLO/images/Riyadh_people.png', cv2.IMREAD_UNCHANGED)

    factor = 4 # downsampling factor 

    # find old and new image dimensions
    h, w, _ = img.shape
    new_height = int(h / factor)
    new_width = int(w / factor)

    # resize the image - down
    img = cv2.resize(img, (new_width, new_height), interpolation = cv2.INTER_LINEAR)

else:

    img = cv2.imread('/home/noumena/Documents/VideoProcessing_improvingYOLO/images/Riyadh_people.png', cv2.IMREAD_UNCHANGED)

```

### **Remove blue channel**

```python
# assign blue channel to zeros
img_blue = img
img_blue[:,:,0] = np.zeros([img_blue.shape[0], img_blue.shape[1]])

#show image
plt.imshow(img_blue)
```

### **Get more vibrant colors**

```python
hsvImg = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

#multiple by a factor to change the saturation
hsvImg[...,1] = hsvImg[...,1]*1.2

#multiple by a factor of less than 1 to reduce the brightness 
hsvImg[...,2] = hsvImg[...,2]*0.9

hsvImg=cv2.cvtColor(hsvImg,cv2.COLOR_HSV2BGR)

#show image
plt.imshow(hsvImg)

#save image
cv2.imwrite('images/hsv_enhanced.jpg', cv2.cvtColor(hsvImg, cv2.COLOR_RGB2BGR)) 
```

### **Gray scale image**

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#show image
plt.imshow(gray)

#save image
cv2.imwrite('images/grayscale.jpg',cv2.cvtColor(gray, cv2.COLOR_RGB2BGR)) 
```
